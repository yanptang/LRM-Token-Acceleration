```python
import json
import pandas as pd
import plotly.graph_objects as go

with open("/data/users/tongf/master_thesis_tang/results/qwen3_1_7b_e2e_100_results.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data["results"])
df = df.sort_values("id").reset_index(drop=True)


import matplotlib.pyplot as plt

for col in ["input_tokens", "output_tokens", "latency_s", "tpot_ms"]:
    plt.figure(figsize=(12,4))
    plt.plot(df["id"], df[col], marker="o")
    plt.xticks(rotation=90)
    plt.title(f"{col} by prompt ID")
    plt.xlabel("Prompt ID")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```
